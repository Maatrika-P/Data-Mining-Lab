Q4.Classification: Decision Trees
Write a program to perform and evaluate classification on a dataset using Decision tree algorithm.

```python
# Load libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import accuracy_score, precision_score, recall_score
import matplotlib.pyplot as plt
```

```python
#Loading data from CSV file
df = pd.read_csv('DM_Lab_Datasets/data.csv')
```

```python
df.head()
```

```python
df.diagnosis[df["diagnosis"]=="B"]=0.0
df.diagnosis[df["diagnosis"]=="M"]=1.0
df['diagnosis'] = df['diagnosis'].astype('int')
y = df['diagnosis']
df.drop("diagnosis", axis=1, inplace=True)
X = df
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=13)
```

```python

Model= DecisionTreeClassifier(max_depth=5)
Model.fit(x_train, y_train)
y_pred = Model.predict(x_test)
#acc= accuracy_score(y_test,y_pred)
print("Accuracy\t:"+str(accuracy_score(y_test,y_pred)))
print("Precision\t:"+str(precision_score(y_test,y_pred)))
print("Recall\t:"+str(recall_score(y_test,y_pred)))
```

```python
from sklearn.metrics import auc, roc_curve

prob=Model.predict_proba(x_test)
prob = prob[:,1]
fpr,tpr,_ = roc_curve(y_test, prob)
roc_auc = auc(fpr, tpr)
print("roc_auc\t:"+str(roc_auc))
```

```python

plt.figure()
lw = 2
plt.plot(
    fpr,
    tpr,
    color="blue",
    lw=lw,
    label="ROC curve (area = %0.2f)" % roc_auc,)
plt.plot([0, 1], [0, 1], color="pink", lw=lw, linestyle="--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic(ROC) example")
plt.legend(loc="lower right")
plt.show()
```
